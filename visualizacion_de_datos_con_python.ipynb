# Visualización de datos con Python: Matplotlib, Seaborn y Plotly

```python
import pandas as pd

df = pd.read_csv('/content/canadian_immegration_data.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
# indica todos los datos y que no estan nulos non null
```

```python
df.set_index('Country', inplace=True)
# Establece la columna 'Country' como índice del DataFrame y modifica el DataFrame original.
df
```

```python
anos = list(map(str, range(1980, 2014)))
anos
# creamos una lista de string de todos los años solicitados
```

```python
colombia = df.loc['Colombia', anos]
# Selecciona la fila donde el índice es 'Colombia' y todas las columnas especificadas en la lista 'anos'.
colombia
# Muestra la Serie resultante con los datos de inmigración de Colombia para cada año en 'anos'.
```

```python
col_dic = {'Año' : colombia.index.tolist(), 'Inmigrantes' : colombia.values.tolist()}
col_dic
# creamos una lista
```

```python
datos_col = pd.DataFrame(col_dic)
datos_col.tail()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 4))
# Crea una nueva figura para el plot con un tamaño de 8x4 pulgadas.

plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
# Genera el plot de línea utilizando los datos de las columnas 'Año' e 'Inmigrantes' del DataFrame datos_col.

plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
# Establece las marcas del eje x en los años especificados.

plt.xlabel('Años')
# Establece la etiqueta del eje x como 'Años'.

plt.ylabel('Inmigrantes')
# Establece la etiqueta del eje y como 'Inmigrantes'.

plt.show()
# Muestra el plot generado.
```

```python
plt.figure(figsize=(8, 4))
# Crea una nueva figura para el plot con un tamaño de 8x4 pulgadas.

plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
# Genera el plot de línea utilizando los datos de las columnas 'Año' e 'Inmigrantes' del DataFrame datos_col.

plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
# Establece las marcas del eje x en los años especificados.

plt.title('Inmigrantes de Colombiahacia canadá')
# Establece el título del plot como 'Inmigrantes de Colombia'.
plt.xlabel('Años')
# Establece la etiqueta del eje x como 'Años'.
plt.ylabel('Numero de Inmigrantes')
# Establece la etiqueta del eje y como 'Inmigrantes'.

plt.show()
# Muestra el plot generado.
```

```python
datos_arg = pd.DataFrame(arg_dic)
datos_bra = pd.DataFrame(bra_dic)
datos_peru = pd.DataFrame(per_dic)
```

```python
plt.figure(figsize=(8, 4))
# Crea una nueva figura para el plot con un tamaño de 8x4 pulgadas.

plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
# Genera el plot de línea utilizando los datos de las columnas 'Año' e 'Inmigrantes' del DataFrame datos_col.
plt.plot(datos_arg['Año'], datos_arg['Inmigrantes'])
plt.plot(datos_bra['Año'], datos_bra['Inmigrantes'])
plt.plot(datos_bra['Año'], datos_peru['Inmigrantes'])


plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
# Establece las marcas del eje x en los años especificados.

plt.title('Inmigrantes de Colombiahacia canadá')
# Establece el título del plot como 'Inmigrantes de Colombia'.
plt.xlabel('Años')
# Establece la etiqueta del eje x como 'Años'.
plt.ylabel('Numero de Inmigrantes')
# Establece la etiqueta del eje y como 'Inmigrantes'.

plt.show()
# Muestra el plot generado.
```

```python
# almacenaremos nuestros graficos en variables
fig, ax = plt.subplots(figsize=(8, 4))
# plt.subplots(): Esta función crea una figura y un conjunto de subtramas.
# figsize=(8, 4): Este argumento establece el tamaño de la figura en 8 pulgadas de ancho por 4 pulgadas de alto.
# fig: Representa la figura completa.
# ax: Representa los ejes (el área donde se dibujará el gráfico).

ax.plot(datos_col['Año'], datos_col['Inmigrantes'])
# ax.plot(): Este método traza los datos en los ejes especificados por 'ax'.
# datos_col['Año']: Utiliza la columna 'Año' del DataFrame datos_col para los valores del eje x.
# datos_col['Inmigrantes']: Utiliza la columna 'Inmigrantes' del DataFrame datos_col para los valores del eje y.

ax.xaxis.set_major_locator(plt.MultipleLocator(5))
# ax.xaxis: Accede al eje x de los ejes 'ax'.
# set_major_locator(): Este método establece la ubicación de las marcas principales en el eje.
# plt.MultipleLocator(5): Crea un localizador que coloca las marcas principales en múltiplos de 5.

ax.set_title('Inmigrantes de Colombianos a canadá \n1980 - 2013')

ax.set_xlabel('Años')
# ax.set_xlabel(): Establece la etiqueta del eje x como 'Años'.

ax.set_ylabel('Numero de Inmigrantes')
# ax.set_ylabel(): Establece la etiqueta del eje y como 'Numero de Inmigrantes'.

plt.show()
# plt.show(): Muestra la figura con el gráfico generado.

# cuando se quiera observar en tiempo se debe usar un grafico de lineas
```

 #### CREANDO SUBPLOTS EN UNA DIRECION

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 3))
# creamos un subplot podemos colocar las filas y la cantidad de columnas, quiero una fila y dos espacios para grafica
# plt.subplots(): Esta función crea una figura y un conjunto de subtramas.
# 1, 2: Especifica el número de filas (1) y columnas (2) para las subtramas.
# figsize=(10, 3): Este argumento establece el tamaño de la figura en 10 pulgadas de ancho por 3 pulgadas de alto.
# fig: Representa la figura completa.
# axs: Representa los ejes (el área donde se dibujará el gráfico) como un arreglo NumPy bidimensional.

axs[0].plot(datos_col['Año'], datos_col['Inmigrantes'])
# axs[0]: Selecciona el primer subplot (el de la izquierda).
# ax.plot(): Este método traza los datos en los ejes especificados por 'ax'.
# datos_col['Año']: Utiliza la columna 'Año' del DataFrame datos_col para los valores del eje x.
# datos_col['Inmigrantes']: Utiliza la columna 'Inmigrantes' del DataFrame datos_col para los valores del eje y.
axs[0].set_title('Inmigracion de colombianos hacia canada\n1980 - 2013')
# Establece el título del primer subplot.
axs[0].set_xlabel('Años')
# Establece la etiqueta del eje x para el primer subplot.
axs[0].set_ylabel('Numero de Inmigrantes')
# Establece la etiqueta del eje y para el primer subplot.
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
# axs[0].xaxis: Accede al eje x del primer subplot.
# set_major_locator(): Este método establece la ubicación de las marcas principales en el eje.
# plt.MultipleLocator(5): Crea un localizador que coloca las marcas principales en múltiplos de 5.
axs[0].grid()


axs[1].boxplot(datos_col['Inmigrantes'])
# axs[1]: Selecciona el segundo subplot (el de la derecha).
# axs[1].boxplot(): Crea un boxplot en el segundo subplot.
# datos_col['Inmigrantes']: Utiliza los datos de la columna 'Inmigrantes' del DataFrame datos_col para crear el boxplot.
axs[1].set_title('Boxplot de la inmigracion de colombianos\nhacia canada 1980 - 2013')
# Establece el título del segundo subplot.
axs[1].set_xlabel('Colombia')
# Establece la etiqueta del eje x para el segundo subplot.
axs[1].set_ylabel('Numero de inmigrantes')
# Establece la etiqueta del eje y para el segundo subplot.
axs[1].grid()

plt.show()
# plt.show(): Muestra la figura con el gráfico generado.

# cuando se quiera observar en tiempo se debe usar un grafico de lineas
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 6))
# Crea una figura y una cuadrícula de subtramas de 2 filas y 2 columnas.
# fig: la figura completa.
# axs: un arreglo NumPy de los ejes de las subtramas.
# figsize=(10, 6): establece el tamaño de la figura en 10 pulgadas de ancho por 6 pulgadas de alto.

fig.subplots_adjust(hspace=0.5, wspace=0.3)
# Ajusta el espacio vertical (hspace) y horizontal (wspace) entre las subtramas.

# configuracion de titulo para todo el lienzo
fig.suptitle('Inmigracion de los 4 mayores paises sudamericanos \n hacia canada 1980 - 2013')
# Establece un título principal para toda la figura.

axs[0,0].plot(df.loc['Colombia',anos])
# Grafica los datos de inmigración de Colombia en la primera subtrama (fila 0, columna 0).
axs[0,0].set_title('Colombia')
# Establece el título para la subtrama de Colombia.

axs[0,1].plot(df.loc['Brazil',anos])
# Grafica los datos de inmigración de Brasil en la segunda subtrama (fila 0, columna 1).
axs[0,1].set_title('Brazil')
# Establece el título para la subtrama de Brasil.

axs[1,0].plot(df.loc['Argentina',anos])
# Grafica los datos de inmigración de Argentina en la tercera subtrama (fila 1, columna 0).
axs[1,0].set_title('Argentina')
# Establece el título para la subtrama de Argentina.

axs[1,1].plot(df.loc['Peru',anos])
# Grafica los datos de inmigración de Perú en la cuarta subtrama (fila 1, columna 1).
axs[1,1].set_title('Peru')
# Establece el título para la subtrama de Perú.

# ajustamos los limites de la medida
ymin = 0
ymax = 7000
for ax in axs.ravel():
    # Itera sobre todas las subtramas en el arreglo axs (aplanado).
    ax.set_ylim(ymin, ymax)
    # Establece los límites del eje y para cada subtrama.

for ax in axs.flat:
    # Itera sobre todas las subtramas en el arreglo axs (aplanado).
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    # Establece los ticks principales del eje x cada 5 unidades para cada subtrama.
    ax.set(xlabel='Años', ylabel='Numero de Inmigrantes')
    # Establece las etiquetas del eje x y del eje y para cada subtrama.
    ax.grid()
    # Agrega una cuadrícula a cada subtrama.

plt.show()
# Muestra la figura con todas las subtramas.
```

```python
# almacenaremos nuestros graficos en variables
fig, ax = plt.subplots(figsize=(8, 4))
# lw hace la lina de tenedencia mas grande
# marker agrega marcadores en punto
ax.plot(datos_col['Año'], datos_col['Inmigrantes'] , lw = 3) #marker= 'o')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
#se cambia el tamaño de la letra
# loc especifica a la izquierda el nombre del titulo
ax.set_title('Inmigrantes de Colombianos a canadá \n1980 - 2013', fontsize = 18, loc ='left')
ax.set_xlabel('Años', fontsize=14)
ax.set_ylabel('Numero de Inmigrantes', fontsize=14)
# cambiar el tamañano de los ejes
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=12)

ax.grid()
plt.show()



# cuando se quiera observar en tiempo se debe usar un grafico de lineas
```

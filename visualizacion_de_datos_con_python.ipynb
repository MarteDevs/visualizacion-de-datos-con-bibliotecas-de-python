```python
import pandas as pd

df = pd.read_csv('/content/canadian_immegration_data.csv')
df.head()
```

```python
df.set_index('Country', inplace=True)
# Establece la columna 'Country' como índice del DataFrame y modifica el DataFrame original.
df
```

```python
anos = list(map(str, range(1980, 2014)))
anos
# creamos una lista de string de todos los años solicitados
```

```python
colombia = df.loc['Colombia', anos]
# Selecciona la fila donde el índice es 'Colombia' y todas las columnas especificadas en la lista 'anos'.
colombia
# Muestra la Serie resultante con los datos de inmigración de Colombia para cada año en 'anos'.
```

```python
# almacenaremos nuestros graficos en variables
fig, ax = plt.subplots(figsize=(8, 4))
# plt.subplots(): Esta función crea una figura y un conjunto de subtramas.
# figsize=(8, 4): Este argumento establece el tamaño de la figura en 8 pulgadas de ancho por 4 pulgadas de alto.
# fig: Representa la figura completa.
# ax: Representa los ejes (el área donde se dibujará el gráfico).

ax.plot(datos_col['Año'], datos_col['Inmigrantes'])
# ax.plot(): Este método traza los datos en los ejes especificados por 'ax'.
# datos_col['Año']: Utiliza la columna 'Año' del DataFrame datos_col para los valores del eje x.
# datos_col['Inmigrantes']: Utiliza la columna 'Inmigrantes' del DataFrame datos_col para los valores del eje y.

ax.xaxis.set_major_locator(plt.MultipleLocator(5))
# ax.xaxis: Accede al eje x de los ejes 'ax'.
# set_major_locator(): Este método establece la ubicación de las marcas principales en el eje.
# plt.MultipleLocator(5): Crea un localizador que coloca las marcas principales en múltiplos de 5.

ax.set_title('Inmigrantes de Colombianos a canadá \n1980 - 2013')

ax.set_xlabel('Años')
# ax.set_xlabel(): Establece la etiqueta del eje x como 'Años'.

ax.set_ylabel('Numero de Inmigrantes')
# ax.set_ylabel(): Establece la etiqueta del eje y como 'Numero de Inmigrantes'.

plt.show()
# plt.show(): Muestra la figura con el gráfico generado.

# cuando se quiera observar en tiempo se debe usar un grafico de lineas
```

 #### CREANDO SUBPLOTS EN UNA DIRECION

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 3))
# creamos un subplot podemos colocar las filas y la cantidad de columnas, quiero una fila y dos espacios para grafica
# plt.subplots(): Esta función crea una figura y un conjunto de subtramas.
# 1, 2: Especifica el número de filas (1) y columnas (2) para las subtramas.
# figsize=(10, 3): Este argumento establece el tamaño de la figura en 10 pulgadas de ancho por 3 pulgadas de alto.
# fig: Representa la figura completa.
# axs: Representa los ejes (el área donde se dibujará el gráfico) como un arreglo NumPy bidimensional.

axs[0].plot(datos_col['Año'], datos_col['Inmigrantes'])
# axs[0]: Selecciona el primer subplot (el de la izquierda).
# ax.plot(): Este método traza los datos en los ejes especificados por 'ax'.
# datos_col['Año']: Utiliza la columna 'Año' del DataFrame datos_col para los valores del eje x.
# datos_col['Inmigrantes']: Utiliza la columna 'Inmigrantes' del DataFrame datos_col para los valores del eje y.
axs[0].set_title('Inmigracion de colombianos hacia canada\n1980 - 2013')
# Establece el título del primer subplot.
axs[0].set_xlabel('Años')
# Establece la etiqueta del eje x para el primer subplot.
axs[0].set_ylabel('Numero de Inmigrantes')
# Establece la etiqueta del eje y para el primer subplot.
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
# axs[0].xaxis: Accede al eje x del primer subplot.
# set_major_locator(): Este método establece la ubicación de las marcas principales en el eje.
# plt.MultipleLocator(5): Crea un localizador que coloca las marcas principales en múltiplos de 5.
axs[0].grid()


axs[1].boxplot(datos_col['Inmigrantes'])
# axs[1]: Selecciona el segundo subplot (el de la derecha).
# axs[1].boxplot(): Crea un boxplot en el segundo subplot.
# datos_col['Inmigrantes']: Utiliza los datos de la columna 'Inmigrantes' del DataFrame datos_col para crear el boxplot.
axs[1].set_title('Boxplot de la inmigracion de colombianos\nhacia canada 1980 - 2013')
# Establece el título del segundo subplot.
axs[1].set_xlabel('Colombia')
# Establece la etiqueta del eje x para el segundo subplot.
axs[1].set_ylabel('Numero de inmigrantes')
# Establece la etiqueta del eje y para el segundo subplot.
axs[1].grid()

plt.show()
# plt.show(): Muestra la figura con el gráfico generado.

# cuando se quiera observar en tiempo se debe usar un grafico de lineas
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 6))
# Crea una figura y una cuadrícula de subtramas de 2 filas y 2 columnas.
# fig: la figura completa.
# axs: un arreglo NumPy de los ejes de las subtramas.
# figsize=(10, 6): establece el tamaño de la figura en 10 pulgadas de ancho por 6 pulgadas de alto.

fig.subplots_adjust(hspace=0.5, wspace=0.3)
# Ajusta el espacio vertical (hspace) y horizontal (wspace) entre las subtramas.

# configuracion de titulo para todo el lienzo
fig.suptitle('Inmigracion de los 4 mayores paises sudamericanos \n hacia canada 1980 - 2013')
# Establece un título principal para toda la figura.

axs[0,0].plot(df.loc['Colombia',anos])
# Grafica los datos de inmigración de Colombia en la primera subtrama (fila 0, columna 0).
axs[0,0].set_title('Colombia')
# Establece el título para la subtrama de Colombia.

axs[0,1].plot(df.loc['Brazil',anos])
# Grafica los datos de inmigración de Brasil en la segunda subtrama (fila 0, columna 1).
axs[0,1].set_title('Brazil')
# Establece el título para la subtrama de Brasil.

axs[1,0].plot(df.loc['Argentina',anos])
# Grafica los datos de inmigración de Argentina en la tercera subtrama (fila 1, columna 0).
axs[1,0].set_title('Argentina')
# Establece el título para la subtrama de Argentina.

axs[1,1].plot(df.loc['Peru',anos])
# Grafica los datos de inmigración de Perú en la cuarta subtrama (fila 1, columna 1).
axs[1,1].set_title('Peru')
# Establece el título para la subtrama de Perú.

# ajustamos los limites de la medida
ymin = 0
ymax = 7000
for ax in axs.ravel():
    # Itera sobre todas las subtramas en el arreglo axs (aplanado).
    ax.set_ylim(ymin, ymax)
    # Establece los límites del eje y para cada subtrama.

for ax in axs.flat:
    # Itera sobre todas las subtramas en el arreglo axs (aplanado).
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    # Establece los ticks principales del eje x cada 5 unidades para cada subtrama.
    ax.set(xlabel='Años', ylabel='Numero de Inmigrantes')
    # Establece las etiquetas del eje x y del eje y para cada subtrama.
    ax.grid()
    # Agrega una cuadrícula a cada subtrama.

plt.show()
# Muestra la figura con todas las subtramas.
```

```python
# personalziando colroes
# podemos usar en conjunto con seabor y matplot
def generar_grafico(palette):
  fig, ax = plt.subplots(figsize=(8, 4))

  ax.set_frame_on(False)
  ax.xaxis.set_visible(False)
  ax.tick_params(axis='y', which = 'both', size = 0)
  # ====================================================

  ax = sns.barplot(data=top_10, x='Total', y=top_10.index, hue=top_10.index, orient='h', palette= palette, legend=False)
  ax.set_title('Top 10 paices con mayor migracion hacia Canada\n1980 - 2013', loc = 'left', fontsize=16)
  ax.set_xlabel('Numero de Inmigrantes', fontsize=14)
  ax.set_ylabel('')

  for i, j in enumerate(top_10['Total']):
    ax.text(j+20, i, str(j), color = 'black',fontsize = 9, ha='left', va='center')

  plt.show()
```

```python
# cracion de un grafico de  barras

def generar_grafico_sudamerica(palette, countries=None):
  fig, ax = plt.subplots(figsize=(8, 4))
  ax.set(xticklabels=[])

  data_to_plot = sudamerica_sorted
  if countries is not None:
      data_to_plot = sudamerica_sorted.loc[countries]

  ax = sns.barplot(data=data_to_plot, x='Total', y=data_to_plot.index, hue=data_to_plot.index, orient='h', palette= palette, legend=False)
  ax.set_title('Top paices de America latina con mayor migracion hacia Canada\n1980 - 2013', loc = 'left', fontsize=16)
  ax.set_xlabel('Numero de Inmigrantes', fontsize=14)
  ax.set_ylabel('')
  sns.despine()

  for i, j in enumerate(data_to_plot['Total']):
    ax.text(j+20, i, str(j), color = 'black',fontsize = 9, ha='left', va='center')

  plt.show()
```

```python
## Creación de un gráfico de líneas de tendencia
import matplotlib.pyplot as plt
import seaborn as sns

# Establece el tema para los gráficos usando seaborn
sns.set_theme()
# Establece la paleta de colores para los gráficos usando seaborn
sns.set_palette('Dark2')

# Crea una figura y un conjunto de subtramas con un tamaño especificado
fig, ax = plt.subplots(figsize=(10, 5))

# Grafica los datos de inmigración de Brasil como un gráfico de líneas
# df.loc['Brazil', anos]: Selecciona los datos de Brasil a lo largo de los años definidos en 'anos'
# label='Brasil': Establece la etiqueta para esta línea en la leyenda
# lw=3: Establece el ancho de la línea en 3
ax = sns.lineplot(df.loc['Brazil', anos], label='Brasil', lw=3)

# Grafica los datos de inmigración de Argentina como un gráfico de líneas
ax = sns.lineplot(df.loc['Argentina', anos], label='Argentina', lw=3)

# Grafica los datos de inmigración de Perú como un gráfico de líneas
ax = sns.lineplot(df.loc['Peru', anos], label='Perú', lw=3)

# Grafica los datos de inmigración de Colombia como un gráfico de líneas
ax = sns.lineplot(df.loc['Colombia', anos], label='Colombia', lw=3)

# Establece el título del gráfico
# loc='left': Alinea el título a la izquierda
# fontsize=20: Establece el tamaño de la fuente del título
ax.set_title('Inmigración de los principales países de América del Sur\nal Canadá de 1980 a 2013', loc='left', fontsize=20)

# Establece la etiqueta para el eje x
ax.set_xlabel('Año', fontsize=14)

# Establece la etiqueta para el eje y
ax.set_ylabel('Número de Inmigrantes', fontsize=14)

# Establece las marcas principales en el eje x para que sean cada 5 unidades
ax.xaxis.set_major_locator(plt.MultipleLocator(5))

# Agrega una leyenda al gráfico
# title='Países': Establece el título de la leyenda
# loc='upper right': Posiciona la leyenda en la esquina superior derecha
# bbox_to_anchor=(1.18, 1.02): Ajusta la posición del cuadro de la leyenda
ax.legend(title='Países', loc='upper right', bbox_to_anchor=(1.18, 1.02))

# Muestra el gráfico
plt.show()
```

```python
import plotly.express as px
# importando plotly
```

```python
fig = px.line(datos_col, x='Año', y='Inmigrantes',
              title='Inmigrantes de Colombianos a Canadá\n1980 - 2013')
fig.update_traces(line_color='forestgreen', line_width = 3)
fig.update_layout(width=800, height = 400, xaxis = {'tickangle':-45},
                  xaxis_title='Año', yaxis_title='Numero de Inmigrantes',
                  font_family='Arial', font_size=12, font_color='gray')
fig.show()
```

```python
sudamerica.drop(['Continent','Region','Total'],axis=1,inplace=True)
sudamerica.head()
```

```python
sudamerica.drop(['DevName'],axis=1,inplace=True)
```

```python
sudamerica = sudamerica.T
sudamerica.head()
```

```python
# Crea un gráfico de líneas interactivo utilizando Plotly Express.
# sudamerica: DataFrame que contiene los datos de inmigración por país a lo largo de los años.
# x=sudamerica.index: Utiliza el índice del DataFrame (los años) para el eje x.
# y=sudamerica.columns: Utiliza los nombres de las columnas (los países) para el eje y, creando una línea separada para cada país.
# color='Country': Asigna un color diferente a cada línea basándose en el nombre del país (la columna 'Country' en el DataFrame transformado).
# markers=True: Agrega marcadores a cada punto de datos en las líneas.
# title='Inmigrantes de sudamericanos a Canadá\n1980 - 2013': Establece el título del gráfico.
fig = px.line(sudamerica, x=sudamerica.index, y=sudamerica.columns,
              title='Inmigrantes de sudamericanos a Canadá\n1980 - 2013', color = 'Country', markers=True)

# Actualiza el diseño del gráfico.
# width=800, height = 400: Establece el tamaño del gráfico en píxeles.
# xaxis = {'tickangle':-45}: Rota las etiquetas del eje x 45 grados para evitar que se superpongan.
# xaxis_title='Año': Establece la etiqueta del eje x.
# yaxis_title='Numero de Inmigrantes': Establece la etiqueta del eje y.
# font_family='Arial', font_size=12, font_color='gray': Configura la fuente y el color del texto.
fig.update_layout(width=800, height = 400, xaxis = {'tickangle':-45},
                  xaxis_title='Año', yaxis_title='Numero de Inmigrantes',
                  font_family='Arial', font_size=12, font_color='gray')

# Muestra el gráfico interactivo.
fig.show()

# genial un grafico de lienas por pais donde cada uno tiene su propio informacion
```

```python
datos_arg['Año'] = datos_arg['Año'].astype(int)
```

```python
import plotly.graph_objs as go

# Criando uma figura
fig = go.Figure()

# Adicionando a linha do gráfico e definindo a espessura da linha
fig.add_trace(
    go.Scatter(x=[datos_col['Año'].iloc[0]], y=[datos_col['Inmigrantes'].iloc[0]], mode='lines', name='Inmigrantes', line=dict(width=4))
)

# Definir la configuración del layout
fig.update_layout(
    title=dict(
        text='<b>Inmigración de Colombianos hacia Canadá en el periodo de 1980 a 2013</b>',
        x=0.12,
        xanchor='left',
        font=dict(size=20)
    ),
    xaxis=dict(range=[1980, 2013], autorange=False, title='<b>Año</b>'),
    yaxis=dict(range=[0, 7000], autorange=False, title='<b>Número de Inmigrantes</b>'),
    updatemenus=[dict(
        type='buttons',
        showactive=False,
        buttons=[dict(
            label='Play',
            method='animate',
            args=[None, {'frame': {'duration': 100, 'redraw': True}, 'fromcurrent': True}]
        )]
    )],
    width=1000,
    height=500
)

# Definir la configuración de la animación
frames = [go.Frame(data=[go.Scatter(x=datos_col['Año'].iloc[:i+1], y=datos_col['Inmigrantes'].iloc[:i+1])]) for i in range(len(datos_col))]
fig.frames = frames

# Mostrar la figura
fig.show()
```

```python
import plotly.graph_objs as go

# Criando uma figura
fig = go.Figure()

# Adicionando as linhas do gráfico e definindo a espessura da linha
fig.add_trace(
    go.Scatter(x=[datos_col['Año'].iloc[0]], y=[datos_col['Inmigrantes'].iloc[0]], mode='lines', name='Colombia', line=dict(width=4))
)
fig.add_trace(
    go.Scatter(x=[datos_arg['Año'].iloc[0]], y=[datos_arg['Inmigrantes'].iloc[0]], mode='lines', name='Argentina', line=dict(width=4))
)


# Definir la configuración del layout
fig.update_layout(
    title=dict(
        text='<b>Inmigración de Colombianos y Argentinos hacia Canadá en el periodo de 1980 a 2013</b>',
        x=0.12,
        xanchor='left',
        font=dict(size=20)
    ),
    xaxis=dict(range=[1980, 2013], autorange=False, title='<b>Año</b>'),
    yaxis=dict(range=[0, 7000], autorange=False, title='<b>Número de Inmigrantes</b>'),
    updatemenus=[dict(
        type='buttons',
        showactive=False,
        buttons=[dict(
            label='Play',
            method='animate',
            args=[None, {'frame': {'duration': 100, 'redraw': True}, 'fromcurrent': True}]
        )]
    )],
    width=1000,
    height=500
)

# Definir la configuración de la animación
frames = [go.Frame(data=[go.Scatter(x=datos_col['Año'].iloc[:i+1], y=datos_col['Inmigrantes'].iloc[:i+1], name='Colombia'),
                          go.Scatter(x=datos_arg['Año'].iloc[:i+1], y=datos_arg['Inmigrantes'].iloc[:i+1], name='Argentina')]) for i in range(len(datos_col))]
fig.frames = frames

# Mostrar la figura
fig.show()
```

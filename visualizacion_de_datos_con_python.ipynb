# Visualización de datos con Python: Matplotlib, Seaborn y Plotly

```python
import pandas as pd

df = pd.read_csv('/content/canadian_immegration_data.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
# indica todos los datos y que no estan nulos non null
```

```python
df.set_index('Country', inplace=True)
# Establece la columna 'Country' como índice del DataFrame y modifica el DataFrame original.
df
```

```python
anos = list(map(str, range(1980, 2014)))
anos
# creamos una lista de string de todos los años solicitados
```

```python
colombia = df.loc['Colombia', anos]
# Selecciona la fila donde el índice es 'Colombia' y todas las columnas especificadas en la lista 'anos'.
colombia
# Muestra la Serie resultante con los datos de inmigración de Colombia para cada año en 'anos'.
```

```python
col_dic = {'Año' : colombia.index.tolist(), 'Inmigrantes' : colombia.values.tolist()}
col_dic
# creamos una lista
```

```python
datos_col = pd.DataFrame(col_dic)
datos_col.tail()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 4))
# Crea una nueva figura para el plot con un tamaño de 8x4 pulgadas.

plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
# Genera el plot de línea utilizando los datos de las columnas 'Año' e 'Inmigrantes' del DataFrame datos_col.

plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
# Establece las marcas del eje x en los años especificados.

plt.xlabel('Años')
# Establece la etiqueta del eje x como 'Años'.

plt.ylabel('Inmigrantes')
# Establece la etiqueta del eje y como 'Inmigrantes'.

plt.show()
# Muestra el plot generado.
```

```python
plt.figure(figsize=(8, 4))
# Crea una nueva figura para el plot con un tamaño de 8x4 pulgadas.

plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
# Genera el plot de línea utilizando los datos de las columnas 'Año' e 'Inmigrantes' del DataFrame datos_col.

plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
# Establece las marcas del eje x en los años especificados.

plt.title('Inmigrantes de Colombiahacia canadá')
# Establece el título del plot como 'Inmigrantes de Colombia'.
plt.xlabel('Años')
# Establece la etiqueta del eje x como 'Años'.
plt.ylabel('Numero de Inmigrantes')
# Establece la etiqueta del eje y como 'Inmigrantes'.

plt.show()
# Muestra el plot generado.
```

```python
argentina = df.loc['Argentina', anos]
brasil = df.loc['Brazil', anos]
peru = df.loc['Peru', anos]
```

```python
arg_dic = {'Año' : argentina.index.tolist(), 'Inmigrantes' : argentina.values.tolist()}
bra_dic = {'Año' : brasil.index.tolist(), 'Inmigrantes' : brasil.values.tolist()}
per_dic = {'Año' : peru.index.tolist(), 'Inmigrantes' : peru.values.tolist()}
```

```python
datos_arg = pd.DataFrame(arg_dic)
datos_bra = pd.DataFrame(bra_dic)
datos_peru = pd.DataFrame(per_dic)
```

```python
plt.figure(figsize=(8, 4))
# Crea una nueva figura para el plot con un tamaño de 8x4 pulgadas.

plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
# Genera el plot de línea utilizando los datos de las columnas 'Año' e 'Inmigrantes' del DataFrame datos_col.
plt.plot(datos_arg['Año'], datos_arg['Inmigrantes'])
plt.plot(datos_bra['Año'], datos_bra['Inmigrantes'])
plt.plot(datos_bra['Año'], datos_peru['Inmigrantes'])


plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
# Establece las marcas del eje x en los años especificados.

plt.title('Inmigrantes de Colombiahacia canadá')
# Establece el título del plot como 'Inmigrantes de Colombia'.
plt.xlabel('Años')
# Establece la etiqueta del eje x como 'Años'.
plt.ylabel('Numero de Inmigrantes')
# Establece la etiqueta del eje y como 'Inmigrantes'.

plt.show()
# Muestra el plot generado.
```

```python
# almacenaremos nuestros graficos en variables
fig, ax = plt.subplots(figsize=(8, 4))
# plt.subplots(): Esta función crea una figura y un conjunto de subtramas.
# figsize=(8, 4): Este argumento establece el tamaño de la figura en 8 pulgadas de ancho por 4 pulgadas de alto.
# fig: Representa la figura completa.
# ax: Representa los ejes (el área donde se dibujará el gráfico).

ax.plot(datos_col['Año'], datos_col['Inmigrantes'])
# ax.plot(): Este método traza los datos en los ejes especificados por 'ax'.
# datos_col['Año']: Utiliza la columna 'Año' del DataFrame datos_col para los valores del eje x.
# datos_col['Inmigrantes']: Utiliza la columna 'Inmigrantes' del DataFrame datos_col para los valores del eje y.

ax.xaxis.set_major_locator(plt.MultipleLocator(5))
# ax.xaxis: Accede al eje x de los ejes 'ax'.
# set_major_locator(): Este método establece la ubicación de las marcas principales en el eje.
# plt.MultipleLocator(5): Crea un localizador que coloca las marcas principales en múltiplos de 5.

ax.set_title('Inmigrantes de Colombianos a canadá \n1980 - 2013')

ax.set_xlabel('Años')
# ax.set_xlabel(): Establece la etiqueta del eje x como 'Años'.

ax.set_ylabel('Numero de Inmigrantes')
# ax.set_ylabel(): Establece la etiqueta del eje y como 'Numero de Inmigrantes'.

plt.show()
# plt.show(): Muestra la figura con el gráfico generado.

# cuando se quiera observar en tiempo se debe usar un grafico de lineas
```
